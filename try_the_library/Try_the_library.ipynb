###import the library


```python
import graph_cordinates.graph_cordinates as gc
# other libraries
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

### Graph of tesla stock price from July 2022 to July 2023


```python
image_path='tesla-stockprice.png'
image=Image.open(image_path)
plt.imshow(image)
plt.axis('off')
```

### Prepare the inputs to the function:


```python
x_range=[2022,2023]  #range of time
y_range=[100,350]    #range of price
background_value=0   # Because the backgrounde is darker than the curve, we should have given 1 otherwise
```

### Call the function with inputs


```python
arr=gc.get_graphcordinates(image_path,x_range,y_range,background_value)
```

###First 5 and last 5 coordinates

```python
arr[:5],arr[-5:]
```

###Verify the result by plotting the points of the array returned by the function and compare with the original graph


```python
image_arr=np.array(image)
r1= image_arr.shape[0]/image_arr.shape[1]
r2= ( y_range[1]-y_range[0] )/ ( x_range[1]-x_range[0] )
r=r1/r2


fig, ax = plt.subplots(dpi=100)
ax.scatter(arr[:,0],arr[:,1],s=0.1)

ax.set_aspect(r)
ax.set_ylim([2022,2023])
ax.set_ylim([100,350])
plt.show()
```
